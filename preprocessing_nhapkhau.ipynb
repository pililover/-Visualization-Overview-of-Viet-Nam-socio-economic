# Preprocess data Nhập khẩu 2024

```python
import pandas as pd
```

## Đọc nhãn các mặt hàng

```python
df_total = pd.DataFrame()
file_path_template = "data/xuat_nhap/nhap_khau_t{}.xlsx"
file_t1 = file_path_template.format(1)
df_t1 = pd.read_excel(file_t1, header=2)
df_total = df_t1.iloc[:, [0, 1, 4]].copy()
```

## Đọc dữ liệu theo các tháng

```python
for month in range(1, 13):
    file = file_path_template.format(month) 
    df = pd.read_excel(file, header=2)
    df = df.drop(columns=['Unnamed: 2', 'Unnamed: 3', 'So với tháng trước (%)', 'Unnamed: 8', 
                              'Cộng dồn đến hết tháng báo cáo', 'Unnamed: 10', 'Unnamed: 11'])
    df = df.drop(df.index[-1]) 
    df_selected = df.iloc[:, [3, 4, 5, 6]]
    df_selected.columns = [
        f"Lượng Tháng {month}",
        f"Trị giá Tháng {month}",
        f"Lượng So với cùng kỳ Tháng {month}",
        f"Trị giá So với cùng kỳ Tháng {month}"
    ]
    df_total = pd.concat([df_total, df_selected], axis=1)
```

```python
df_total = df_total.drop(df_total.index[-1])
df_total = df_total.drop(df_total.index[0])
```

## Cấu trúc lại các cột

```python
multi_columns = [("Thông tin chung", "STT"),
                 ("Thông tin chung", "Nhóm/Mặt hàng chủ yếu"),
                 ("Thông tin chung", "ĐVT")]

for month in range(1, 13):
    multi_columns.extend([
        (f"Tháng {month}", "Lượng"),
        (f"Tháng {month}", "Trị giá"),
        (f"Tháng {month}", "Lượng So Kỳ"),
        (f"Tháng {month}", "Trị giá So Kỳ")
    ])

df_total.columns = pd.MultiIndex.from_tuples(multi_columns)
df_total
```

## Lưu data

```python
output_file = "data/xuat_nhap/nhap_khau_2024.xlsx"
df_total.to_excel(output_file)
```
