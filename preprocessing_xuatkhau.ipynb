```python
import pandas as pd
```

```python
columns = ["STT", "Nhóm/Mặt hàng chủ yếu", "ĐVT"]
categories = ["Số trong tháng báo cáo", "So với cùng kỳ"]
data = {category: pd.DataFrame() for category in categories}
```

```python
for month in range(1, 13):
    file_path = f"data/xuat_nhap/xuat_khau_t{month}.xlsx"
    df = pd.read_excel(file_path, header=2)
    df = df.drop(columns=['Unnamed: 3', 'So với tháng trước (%)', 'Unnamed: 8', 
                          'Cộng dồn đến hết tháng báo cáo', 'Unnamed: 10', 'Unnamed: 11'],
                 errors='ignore')  
    df = df.iloc[3:-1]
    if month == 1:
        for category in categories:
            data[category] = df[columns].copy()
    col_index = 4 + 0
    for i, category in enumerate(categories):
        if col_index < len(df.columns):  
            data[category][f'Lượng Tháng {month}'] = df.iloc[:, col_index]
            data[category][f"Trị giá tháng {month}"] = df.iloc[:, col_index + 1]
        col_index = col_index + 2 
```

```python
output_file = "data/xuat_nhap/xuat_khau_2024.xlsx"
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:
    for sheet_name, df_sheet in data.items():
        df_sheet.to_excel(writer, sheet_name=sheet_name, index=False)

print(data)
```
